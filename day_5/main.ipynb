```python
#highscore challenge
# Input a list of student scores
student_scores = input().split()
for n in range(0, len(student_scores)):
  student_scores[n] = int(student_scores[n])

# Write your code below this row üëá
highest_score = 0
for score in student_scores:
  if score > highest_score:
    highest_score = score

print(f"The highest score in the class is: {highest_score}")
```

```python
#average height
# Input a Python list of student heights
student_heights = input().split()
for n in range(0, len(student_heights)):
  student_heights[n] = int(student_heights[n])
# üö® Don't change the code above üëÜ
  
# Write your code below this row üëá
count = 0
sh = 0
for h in student_heights:
  count += 1
  sh = sh + h
average = sh/count

print(f'total height = {sh}')
print(f'number of students = {count}')
print(f'average height = {round(average)}')
```

```python
#fizzbuzz challenge
# print fizz if divisible by 3 instead of printing number
# print buzz if divisible by 5 instead of printing number 
# print fizzbuzz if divisible by both 3 and 5
# Write your code here üëá
target = 100

for i in range(1, target+1):
    if i % 3 == 0:
      if i % 5 == 0:
        print('FizzBuzz')
      else:
        print('Fizz')
    elif i % 5 == 0:
      if i % 5 == 3:
        print('FizzBuzz')
      else:
        print('Buzz')
    else:
      print(i)
```

```python
#shorter version (professors solution)
target = 100
for number in range(1, target + 1):
  if number % 3 == 0 and number % 5 == 0:
    print("FizzBuzz")
  elif number % 3 == 0:
    print("Fizz")
  elif number % 5 == 0:
    print("Buzz")
  else:
    print(number)
```
